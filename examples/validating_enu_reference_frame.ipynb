```python
import diodem
import qmt
import matplotlib.pyplot as plt
```

```python
segment = "seg2"
data = diodem.load_data(1, 3, 3, resample_to_hz=10.0)[segment]

quat, acc, mag = data["quat"], data["imu_rigid"]["acc"], data["imu_rigid"]["mag"]
# rotate from sensor frame to earth reference frame
acc_eps, mag_eps = qmt.rotate(quat, acc), qmt.rotate(quat, mag)
```

```python
colors = ["C0", "C1", "C2"]
labels = ["x", "y", "z"]

for i in range(3):
    plt.plot(mag[:, i], label=labels[i] + "_sensor_frame", color=colors[i])
    plt.plot(mag_eps[:, i], label=labels[i] + "_earth_frame", color=colors[i], linestyle="dashed")
plt.legend()
```

```python
colors = ["C0", "C1", "C2"]
labels = ["x", "y", "z"]

for i in range(3):
    plt.plot(acc[:, i], label=labels[i] + "_sensor_frame", color=colors[i])
    plt.plot(acc_eps[:, i], label=labels[i] + "_earth_frame", color=colors[i], linestyle="dashed")
plt.legend()
```
